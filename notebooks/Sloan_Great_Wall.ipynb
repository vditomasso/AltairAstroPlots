```python
# Author: Jake VanderPlas <vanderplas@astro.washington.edu>
# License: BSD
#   The figure is an example from astroML: see http://astroML.github.com
# http://www.astroml.org/examples/datasets/plot_great_wall.html
import numpy as np
from matplotlib import pyplot as plt

from astroML.datasets import fetch_great_wall
from astroML.density_estimation import KDE, KNeighborsDensity

#------------------------------------------------------------
# Fetch the great wall data
X = fetch_great_wall()

print('X=',X)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(8, 4))

# First plot: scatter the points
ax = plt.subplot(111, aspect='equal')
ax.scatter(X[:, 1], X[:, 0], s=1, lw=0, c='k')

x_axis_data = X[:, 1]
y_axis_data = X[:, 0]

print("X[:, 0] = ",X[:, 0])
print(type(X[:, 0]))

ax.set_xlim(-300, 200)
ax.set_ylim(-375, -175)

ax.set_xlabel('y (Mpc)')
ax.set_ylabel('x (MPC)')

plt.show()
```

```python
# Put your data into a pandas dataframe

import pandas as pd

from astroML.datasets import fetch_great_wall
from astroML.density_estimation import KDE, KNeighborsDensity

X = fetch_great_wall

data = pd.DataFrame({'y (Mpc)': list(x_axis_data),
                     'x (MPC)': list(y_axis_data)})
```

```python
# Import altair
from altair import *

# Set the data you'll be plotting, the color and size of your points
chart = Chart(data).mark_point(color='black', size=0.004).encode(x='y (Mpc)', y='x (MPC)')

# Set your chart so that you can plot all of your data
chart.max_rows=8050

# Set your y-axis
chart.encode(Y('x (MPC)', scale=Scale(domain=[-375, -175])))

# Turn off the underlying grid
chart.configure_axis(grid=False)

# Set the size of your figure
chart.configure_cell(height=175,width=450)
```
